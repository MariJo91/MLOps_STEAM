1. Cargar El Dataset Australian User Reviews asignado por SoyHenry - Conjuntos de Datos de la Plataforma Steam

```python
def load_json_lines(file_path):
    """Load a JSON Lines file."""
    data = []
    with open(file_path, 'r', encoding = 'utf-8 ') as file:
        for line in file:
            data.append(ast.literal_eval(line))
    return pd.DataFrame(data)
```

```python
# Cargar y luego mostrar el Archivo Australian User Reviews Json
df_reviews = load_json_lines("australian_user_reviews.json")
```

```python
# Corregir el formato del JSON
datos_reviews= df_reviews.replace("'", '"')
```

La Columna Reviews, esta Anidada. Debemos transformarla.

```python
def Eliminar_duplicados(df_reviews):
    #Utilizamos json normalize para aplanar los valores en la columna
    flattened_reviews = json_normalize(df_reviews["reviews"])

    #Agregamos user_id de primero
    flattened_reviews = pd.concat([df_reviews["user_id"], flattened_reviews], axis=1)

    #Eliminamos las filas que no tienen, Valores duplicados
    user_reviews = flattened_reviews[flattened_reviews.duplicated(subset = "user_id")]
    return user_reviews
```

```python
user_reviews = Eliminar_duplicados(df_reviews)

user_reviews
```

```python
user_reviews = pd.melt(user_reviews, id_vars=["user_id"], 
                       value_vars=user_reviews.columns[2:11],
                       value_name="reviews")
user_reviews.head()
```

```python
df_reviews["reviews_posted"] = df_reviews["reviews_posted"].astype(str).str.replace(r"Posted |,|/", "", regex=True)

# Convertir la cadena a datetime
df_reviews["reviews_posted"] = pd.to_datetime(df_reviews["reviews_posted"], errors='coerce')

# Ahora df_reviews["reviews_posted"] debería ser de tipo datetime
```

```python
#Creamos la Coluna Year
df_reviews["year"] = df_reviews["reviews_posted"].dt.year.astype("Int64")
#Ordenamos el DataFrame
ddf_reviews = df_reviews.sort_values(["reviews_item_id", "year"])
```

```python
#Rellenar Valores Nulos
ddf_reviews["year"] = ddf_reviews.groupby("reviews_item_id", group_keys=False)["year"].apply(lambda group: group.interpolate())
```

```python
#Verificamos si quedan valores nulos, luego de la interpolacion y los rellenamos
ddf_reviews["year"] = ddf_reviews["year"].fillna(ddf_reviews["year"].median())
```

```python
#Evaluamos el lenguaje de la columna review
def detectar_idioma(texto):
    try:
        return detect(texto)
    except:
        return None
```
